```python
#Scrapper Apollo 5th step Step enrich companies CSV

import pandas as pd
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.options import Options
import time

try:
    # Leer el archivo CSV
    df = pd.read_csv('Resultados/Unidos/Apollo_Acounts_Full copy.csv')

    # Crear nuevas columnas 'Website', 'LinkedIn URL', 'Facebook URL', 'Twitter URL', 'Descripcion' y 'Technologies'
    df['Website'] = ''
    df['LinkedIn URL'] = ''
    df['Facebook URL'] = ''
    df['Twitter URL'] = ''
    df['Descripcion'] = ''
    df['Technologies'] = ''

    # Iterar sobre cada fila del DataFrame
    for index, row in df.iterrows():
        company_url = row['Company URL']

        def scrape_data(attempt=1):
            if attempt > 2:  # Si ya se intentó 2 veces, saltar a la siguiente fila
                print(f"Se ha intentado 2 veces sin éxito para la fila {index}. Continuando con la siguiente fila.")
                return

            try:
                # Navegar al Company URL
                driver.get(company_url)

                # Esperar hasta que se encuentre el div que contiene los enlaces
                links_div = WebDriverWait(driver, 30).until(
                    EC.presence_of_element_located((By.XPATH, '/html/body/div[2]/div/div[2]/div[2]/div/div[2]/div/div/div[2]/div[2]/div/div/div/div[1]/div/div/div/div/div[2]'))
                )

                # Encontrar todos los enlaces dentro del div
                links = links_div.find_elements(By.TAG_NAME, 'a')

                # Inicializar variables para almacenar los enlaces encontrados
                website_url = ''
                linkedin_url = ''
                facebook_url = ''
                twitter_url = ''

                # Iterar sobre los enlaces encontrados y asignar los URLs a las variables correspondientes
                for link in links:
                    href = link.get_attribute('href')
                    if (href.startswith("https://www.") or href.startswith("http://www.")) and not (
                            href.startswith("https://www.linkedin.com/") or href.startswith("http://www.linkedin.com/") or
                            href.startswith("https://www.facebook.com/") or href.startswith("http://www.facebook.com/") or
                            href.startswith("https://twitter.com/") or href.startswith("http://twitter.com/") or
                            href.startswith("https://x.com/") or href.startswith("http://x.com/")):
                        website_url = href
                    if href.startswith("https://www.linkedin.com/") or href.startswith("http://www.linkedin.com/"):
                        linkedin_url = href
                    if href.startswith("https://www.facebook.com/") or href.startswith("http://www.facebook.com/"):
                        facebook_url = href
                    if href.startswith("https://twitter.com/") or href.startswith("http://twitter.com/"):
                        twitter_url = href

                # Asignar los valores encontrados a las columnas correspondientes en el DataFrame
                df.at[index, 'Website'] = website_url
                df.at[index, 'LinkedIn URL'] = linkedin_url
                df.at[index, 'Facebook URL'] = facebook_url
                df.at[index, 'Twitter URL'] = twitter_url

                # Hacer clic en el enlace antes de scrapear la descripción
                try:
                    description_link = WebDriverWait(driver, 20).until(
                        EC.presence_of_element_located((By.XPATH, '/html/body/div[2]/div/div[2]/div[2]/div/div[2]/div/div/div[2]/div[2]/div/div/div/div[2]/div/div[2]/div/div[1]/div[1]/div/div/div/div[1]/div[1]/div/div/div/span/a'))
                    )
                    description_link.click()
                    time.sleep(3)  # Esperar 3 segundos para asegurar que la página se cargue después del clic

                    # Esperar hasta que se encuentre el elemento de la descripción y obtener el texto
                    descripcion_elem = WebDriverWait(driver, 20).until(
                        EC.presence_of_element_located((By.XPATH, '/html/body/div[2]/div/div[2]/div[2]/div/div[2]/div/div/div[2]/div[2]/div/div/div/div[2]/div/div[2]/div/div[1]/div[1]/div/div/div/div[1]/div[1]/div/div/div/span/span'))
                    )
                    descripcion_text = descripcion_elem.text

                except Exception as e:
                    descripcion_text = ''
                    print(f"No se pudo hacer clic en el enlace de descripción o obtener la descripción para la fila {index}")

                # Asignar la descripción encontrada a la columna correspondiente en el DataFrame
                df.at[index, 'Descripcion'] = descripcion_text

                # Hacer clic en el enlace con la palabra "Technologies" antes de scrapear las tecnologías
                try:
                    technologies_link = WebDriverWait(driver, 20).until(
                        EC.element_to_be_clickable((By.XPATH, '/html/body/div[2]/div/div[2]/div[2]/div/div[2]/div/div/div[2]/div[2]/div/div/div/div[2]/div/div[2]/div/div[1]/div[4]/div/div/div/div/div[2]/div/div/div[1]/div/a[2]/div[2]'))
                    )
                    technologies_link.click()
                    time.sleep(3)  # Esperar 3 segundos para asegurar que la página se cargue después del clic

                    # Esperar hasta que se encuentre el elemento de tecnologías y obtener el texto
                    technologies_elem = WebDriverWait(driver, 20).until(
                        EC.presence_of_element_located((By.XPATH, '/html/body/div[2]/div/div[2]/div[2]/div/div[2]/div/div/div[2]/div[2]/div/div/div/div[2]/div/div[2]/div/div[1]/div[4]/div/div/div/div/div[2]/div/div/div[3]/div/div/div'))
                    )
                    technologies_text = technologies_elem.text

                except Exception as e:
                    technologies_text = ''
                    print(f"No se pudo hacer clic en el enlace de 'Technologies' o obtener las tecnologías para la fila {index}")

                # Asignar las tecnologías encontradas a la columna correspondiente en el DataFrame
                df.at[index, 'Technologies'] = technologies_text

                print(f"Datos encontrados para la fila {index}: Website: {website_url}, LinkedIn: {linkedin_url}, Facebook: {facebook_url}, Twitter: {twitter_url}, Descripcion: {descripcion_text}, Technologies: {technologies_text}")

            except Exception as e:
                print(f"No se pudieron obtener los datos para la fila {index}, intentando nuevamente.")
                driver.refresh()
                time.sleep(5)  # Esperar 5 segundos después de refrescar
                scrape_data(attempt + 1)  # Intentar de nuevo

        scrape_data()

    # Guardar el DataFrame actualizado como un archivo CSV
    df.to_csv('100companies.csv', index=False)
    print("CSV actualizado exitosamente con URLs de Website, LinkedIn, Facebook, Twitter, Descripcion y Technologies.")

except Exception as e:
    print(f"No se pudo leer o procesar el archivo CSV o hubo un problema con Selenium")

finally:
    driver.quit()
    print("Proceso de búsqueda de URLs, Descripcion y Technologies completado.")
```

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import time

# Leer el archivo CSV
input_file = '100companies.csv'  # Cambia esto al nombre de tu archivo CSV
df = pd.read_csv(input_file)

# Lista para almacenar los datos extraídos
all_data = []

def process_company_row(row):
    company_url = row['Company URL']
    company_name = row['Company']
    website = row['Website']
    
    def extract_table_data():
        # Buscar el enlace de Employees con el XPath especificado
        employees_link = WebDriverWait(driver, 20).until(
            EC.presence_of_element_located((By.XPATH, '/html/body/div[2]/div/div[2]/div[2]/div/div[2]/div/div/div[2]/div[2]/div/div/div/div[2]/div/div[1]/div/a[5]'))
        )
        employees_link.click()

        # Buscar la tabla con el XPath especificado
        table = WebDriverWait(driver, 20).until(
            EC.presence_of_element_located((By.XPATH, '/html/body/div[2]/div/div[2]/div[2]/div/div[2]/div/div/div[2]/div[2]/div/div/div/div[2]/div/div[2]/div/div/div/div/div/div[4]/div/div/div/div/div[2]/div/table'))
        )

        # Extraer los datos de la tabla
        time.sleep(2)
        table_rows = table.find_elements(By.TAG_NAME, 'tr')
        table_data = []

        # Obtener los encabezados de la tabla
        headers = [header.text for header in table_rows[0].find_elements(By.TAG_NAME, 'th')]
        if not headers:  # Si no hay elementos <th>, usar <td> de la primera fila
            headers = [header.text for header in table_rows[0].find_elements(By.TAG_NAME, 'td')]

        # Filtrar el índice de la columna "Quick Actions"
        try:
            quick_actions_index = headers.index('Quick Actions')
        except ValueError:
            quick_actions_index = -1

        # Eliminar la columna "Quick Actions" de los encabezados si existe
        if quick_actions_index != -1:
            headers.pop(quick_actions_index)

        # Iterar sobre cada fila de la tabla
        for row in table_rows[1:]:  # Omitir la primera fila de encabezados
            cells = row.find_elements(By.TAG_NAME, 'td')
            row_data = [cell.text for i, cell in enumerate(cells) if i != quick_actions_index]

            # Extraer el URL del perfil en la columna "Name"
            if 'Name' in headers:
                name_cell = cells[headers.index('Name')]
                name_link_element = name_cell.find_element(By.TAG_NAME, 'a')
                apollo_profile_url = name_link_element.get_attribute('href')
                row_data.append(apollo_profile_url)
            else:
                row_data.append(None)

            table_data.append(row_data)

        # Agregar el título de la nueva columna "Apollo Profile URL"
        headers.append('Apollo Profile URL')

        # Crear un DataFrame con los datos de la tabla
        table_df = pd.DataFrame(table_data, columns=headers)

        # Agregar las columnas 'Company' y 'Website' a cada fila del DataFrame
        table_df['Company'] = company_name
        table_df['Website'] = website

        # Añadir los datos extraídos a la lista general
        all_data.append(table_df)
    
    try:
        driver.get(company_url)
        extract_table_data()
        
    except Exception as e:
        print(f"Error en la primera tentativa para {company_name}. Reintentando...")
        try:
            driver.refresh()
            extract_table_data()
        except Exception as e2:
            print(f"No se pudo completar la tarea para {company_name} en el segundo intento: {str(e2)}")

# Iterar sobre cada fila del DataFrame
for index, row in df.iterrows():
    print(f"Procesando fila {index + 1} de {len(df)}: {row['Company']}")  # Mensaje para imprimir la fila actual
    process_company_row(row)

# Concatenar todos los DataFrames de la lista en uno solo
if all_data:
    final_df = pd.concat(all_data, ignore_index=True)
    # Guardar el DataFrame final en un archivo CSV
    final_df.to_csv('Prospects_with_all_data.csv', index=False)
    print("Archivo 'Prospects_with_all_data.csv' creado exitosamente.")

print("Proceso terminado")
```
